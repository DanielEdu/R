sacar la raiz cuadrada para afinar la curva.
a la predicciÃ³n hay quie elevar al cuadrado apra desacer lo anterior

```r
tiempo_sqrt <- sqrt(datos$tiempo)
datos['tiempo_sqrt'] <- tiempo_sqrt

head(datos)
```

```r
plot <- ggplot(
    datos, aes(x=unidades, y=tiempo_sqrt)
              ) + geom_point()  + geom_smooth(method = "lm", se = FALSE)
#pintar regresion con ggplot2
plot 
```

```r
# implementar modelo de regreciÃ³n

regresion_sqrt <- lm(tiempo_sqrt ~ unidades, data = datos)
```

```r
plot(regresion_sqrt)
```
